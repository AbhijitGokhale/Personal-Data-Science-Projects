```python
data = mystr

from bs4 import BeautifulSoup

soup = BeautifulSoup(data, "html.parser")
soup
```

```python
parsed_text = [p.get_text() for p in soup.find_all("p", text=True)]
parsed_text
```

```python
string_parsed_text = ''.join(parsed_text)
```

```python
string_parsed_text.split()
```

```python
word_lists = []

    #Loop over all the tweets and extract words into word_list
for i in string_parsed_text.split():
    word_lists.append(i)

word_lists
```

```python
import pandas as pd

def gen_freq1(text1):
    #Will store the list of words
    word_list1 = []

    #Loop over all the tweets and extract words into word_list
    for tw_words1 in text1.split():
        word_list1.append(tw_words1)

    #Create word frequencies using word_list
    word_freq1 = pd.Series(word_list1).value_counts()

    #Print top 20 words
    word_freq1[:20]
    
    return word_freq1

word_freq1_ = gen_freq1(string_parsed_text)
word_freq1_
```

```python
#Import libraries
import matplotlib.pyplot as plt
from wordcloud import WordCloud

#Generate word cloud
wc1 = WordCloud(width=400, height=330, max_words=100, background_color='white').generate_from_frequencies(word_freq1_)

plt.figure(figsize=(12, 8))
plt.imshow(wc1, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# import list of stopwords
from wordcloud import STOPWORDS

print(STOPWORDS)
```

```python
# removing stopwords
word_freq1_ = word_freq1_.drop(labels=STOPWORDS, errors='ignore')

#Generate word cloud
wc1 = WordCloud(width=400, height=330, max_words=100, background_color='white').generate_from_frequencies(word_freq1_)

plt.figure(figsize=(12, 8))
plt.imshow(wc1, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import re

def clean_text(text):
    #Remove will
    #text = re.sub(r'will', '', text)
    
    #Fix &
    #text = re.sub(r'&amp;', '&', text)
    
    #Remove punctuations
    #text = re.sub(r'[?!.;:,#@-]', '', text)
    text = re.sub(r'''[?!.;:'"“”’,#@-]''','',text)

    #Convert to lowercase to maintain consistency
    text = text.lower()
    return text
```

```python
text = clean_text(string_parsed_text)
word_freq1_ = gen_freq1(text)*100
word_freq1_ = word_freq1_.drop(labels=STOPWORDS, errors='ignore')

#Generate word cloud
wc2 = WordCloud(width=400, height=330, max_words=100, background_color='white').generate_from_frequencies(word_freq1_)

plt.figure(figsize=(12, 8))
plt.imshow(wc2, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
data_cleaned2 = data_cleaned1.drop(['host_id','latitude','longitude'],axis = 1)
```

```python
# using OLS to build the linear regression model
from statsmodels.formula.api import ols

lm_out = ols('price~minimum_nights+number_of_reviews+reviews_per_month+calculated_host_listings_count+availability_365', data= data_cleaned2).fit()

# Checking the model results
lm_out.summary()
             
```

```python
# As per above model results, all the independet variables are significant as the respective t value are greater than 3. 
# The 95% CI shows us that higher and lower boundaries does not include 0 which is a good indication that we reject the The 
# NULL hypothesis. The R squared and adjusted R squared value is very low i.e. 1.3% which suggest that the model does not 
# suggest the underlying variablity  properly.
```

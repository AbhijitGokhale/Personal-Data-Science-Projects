```python
from statsmodels.formula.api import ols

lm = ols('charges ~ C(sex) + C(smoker) + C(region) + age + bmi + children', data=data).fit() 

lm.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

variables = lm.model.exog
vif = [variance_inflation_factor(variables, i) for i in range(variables.shape[1])]
vif 
```

The above vif values shows that there is no multicolinearity among independent variables.

```python
lm.model.exog[0]
```

```python
variables
```

```python
np.array(vif).mean()
```

```python
import scipy.stats as stats

#perform the Chi-Square Test of Independence
stats.chi2_contingency(lm.model.exog)
```

Since the p-value (0.823522) of the test is not less than 0.05, we fail to reject the null hypothesis. This means we do not have sufficient evidence to say that there is an association between independent variables.

In other words, the independent variables of the model are independent.

As we can see from the results of multiple linear regression the sex male and region northwest are not significant to the linear model as the p value > 0.05.

```python
# Dependent and Independent variables

y = data['charges']
X = data.drop(['charges'], axis = 1)

from sklearn.model_selection import train_test_split

train_data,test_data = train_test_split(new_data, test_size = 0.2, random_state = 111)
```

```python
train_data
```

```python
test_data
```

```python
from statsmodels.formula.api import ols
lm_n = ols('charges ~ C(sex) + age + bmi + C(smoker) + C(region) + children', data = train_data).fit()
lm_n.summary()
```

```python
lm_n.params
```

```python
y_pred = lm_n.predict(exog = test_data)

y_pred
```
